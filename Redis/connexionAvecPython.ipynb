# Connexion à Redis avec Python

```python
import redis

# Connexion à Redis
r = redis.Redis(host='localhost', port=6379, db=0)

# Vérif de la co
try:
    r.ping()
    print("Connexion réussie : True")
except redis.ConnectionError:
    print("Erreur de connexion à Redis")
```



# Ajout de clés à Redis

```python
r.set('user:1:name', 'John Doe')
r.set('user:1:email', 'john.doe@example.com')

# Clé avec expiration
r.setex('session_key', 60, 'session_data')

print("Données ajoutées à Redis")
```



# Récupération des clés depuis Redis

```python
user_name = r.get('user:1:name').decode('utf-8')
user_email = r.get('user:1:email').decode('utf-8')

print(f"Nom: {user_name}")
print(f"Email: {user_email}")
```



# Récupération multiple de clés

```python
keys = ['user:1:name', 'user:1:email']
values = r.mget(keys)

# Décodage des valeurs
values = [value.decode('utf-8') for value in values]

print("Valeurs récupérées :", values)
```



# Vérification de l'expiration d'une clé

```python
ttl = r.ttl('session_key')
print(f"Temps restant pour 'session_key' : {ttl} secondes")
```
